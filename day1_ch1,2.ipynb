```
# -*- coding: utf-8 -*-
import pandas as pd
!pip install fix_yahoo_finance
import fix_yahoo_finance as yf
import datetime
pd.core.common.is_list_like = pd.api.types.is_list_like
!pip install pandas-datareader
import pandas_datareader.data as pdr
yf.pdr_override()
```

```
# 주식 데이터 다운로드 method

def download_stock_data(file_name, company_code, year1, month1, date1, year2, month2, date2):
    start = datetime.datetime(year1, month1, date1)
    end = datetime.datetime(year2, month2, date2)
    df = pdr.get_data_yahoo(company_code + ".KS", start = start, end = end)

    df.to_pickle(file_name)

    return df

download_stock_data('samsung.data', '005930', 2018,1,1,2018,8,3)
```

### 저장한 주식데이터를 파일로 만들고, 그것을 다시 가져와 load한다.

```
def load_stock_data(file_name):
    df = pd.read_pickle(file_name)
    return df

df = load_stock_data('samsung.data')
print (df.describe())
```

### count : 데이터의 개수
### mean : 데이터의 평균값으로, 산술평균을 의미한다.
### min & max : 데이터에서 제일 작은 값 & 제일 큰 값
### std : standard deviation의 약자로, 표준 편차를 의미한다.

## 사분위수

#### 사분위수는 데이터를 동일한 크기로 4등분, 즉 25%로 4번 나눈것.

사분위는 pandas의 quantile() 함수를 사용하면 쉽게 계산이 가능하다.

```
df = load_stock_data('samsung.data')
print (df.quantile([.25, .5, .45]))
```

## IQR

#### IQR은 분포에서 하위 25%와 상위 25%를 제외한 중간에 있는 50%의 범위를 나타낸다. 

#### IQR = Q3 - Q1

#### IQR은 전체 분포중에서 가운데에 위치하고 있어 대다수 데이터가 어느 범위에 있는지를 가늠할 수 있게 한다.

## 히스토그램

#### 데이터의 구성과 패턴을 파악하는데 많이 사용하는 방법중 하나로, 데이터의 분포를 그래프 형태로 나타낸 것이다.

#### 히스토그램을 그리려면, Bin과 빈도가 필요하다. 
#### Bin은 전체 데이터를 겹치지 않는 일정크리고 나눈 간격을 의미, 빈도는 선택한 Bin에 속한 데이터의 수를 나타낸다.

#### 삼성전자 주가의 히스토그램을 그려보자

```
import matplotlib.pyplot as plt

df = load_stock_data('samsung.data')
(n, bins, patched) = plt.hist(df['Open'])
plt.axvline(df['Open'].mean(), color = 'red')
plt.show()
for index in range(len(n)):
    print " Bin :  %0.f, Frequency = %0.f" % (bins[index], n[index])
```

##히스토그램의 이해

### 히스토그램의 주요항목은 다음과 같다.

* 중심성향 :  데이터가 평균값을 중심으로 분포되어있는가?
* Modes : 데이터 분포상 하나 이상의 무리가 있는가?
* spread : 데이터가 어느 정도로 분산되어 있는가?
* Tail : 하위 25%와 상위 25% 데이터 분포의 기울기 하락도가 완만한가? 아니면 급한가?
* 이상치 : 예외값이 분포도에 존재하는가?

## 정규분포

### : 데이터의 중심을 나타내는 평균과 데이터의 밀집정도를 나타내는 표준편차에 의해 결정되는 분포
### 다른이름으로는 가우스 분포라고 한다. 
### 특히, 평균이 0이고 표준편차가 1인 정규분포를 표준정규분포라고 한다.

## 산점도

### 2개의 변수를 직교좌표계에 그린 그래프
### 변수간의 관계 파악에 유용

### 산점도 행렬

### pandas에서 다수의 변수에 대한 산점도를 한 번에 그려주는 scatter_matrix()라는 함수를 사용한다.

```
from pandas.plotting import scatter_matrix

df = load_stock_data('samsung.data')
scatter_matrix(df[['Open', 'High', 'Low', 'Close']], alpha = 0.2 , figsize =(6,6), diagonal ='kde')
plt.show()
```


 * Open은 커널밀도추정 그래프다. 커널밀도추정은 커널함수를 이요해 어떤 변수의 분포특성을 추정하는것.

## BoxPlot(상자 그림)

### 중앙값, 평균값, 사분위, 이상치 값들을 한눈에 알아볼 수 있는 그래프이다.
### 데이터의 전체 모습을 파악하는 데 매우 유용하게 사용할 수 있다.

## 상자그림의 구성요소

* 이상치 : 양단의 Whisker에 속하지 않은 값으로 일반적인 데이터값이 아닌 값들이다.
* Whisker : 하단의 Whisker는 'Q1 - 1.5 x IQR', 상단의 Whisker는 'Q1 + 1.5 x IQR'로 계산한 값이다.
* Mean : 평균값
* Median : 중앙값
* Quartile : 25th 백분위수는 사분위의 Q1이고 75th 백분위수는 Q3이다.

```
# 상자그림 그리기

df = load_stock_data('samsung.data')
df[['Open', 'High', 'Low', 'Close', 'Adj Close']].plot(kind='box')
plt.show()
```
